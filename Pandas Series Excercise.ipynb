# Series manipulation

```python
#11. Convert the integer Series to float
X = pd.Series([1,2,3,4,5],
              index=['first','second','third','forth','fifth'])
pd.Series(X, dtype=float)   # or X.astype(float)
```

```python
#12. Reverse the Series (first becomes last)
X[::-1]
```

```python
#sort the Series
X.sort_values()
```

```python
#14. Set the fifth element equal to 10
X['fifth'] = 10
```

```python
#15. Change all middle elements to 0
X = pd.Series([1,2,3,4,5],index=['A','B','C','D','E'])
X[1:-1] = 0
X
```

```python
#16. Add 5 to every element
X + 5
```

```python
#17. Create a mask showing negative elements
X = pd.Series([-1,2,0,-4,5,6,0,0,-9,10])
mask = X <= 0
mask
```

```python
#18. Get the negative elements
X = pd.Series([-1,2,0,-4,5,6,0,0,-9,10])
mask = X <= 0
X[mask]
# or directly: X[X <= 0]
```
